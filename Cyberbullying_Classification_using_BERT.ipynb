**Pre-Processing text**

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def preprocess_text(text):
    # Lowercasing
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)

    #tokenize
    tokens = word_tokenize(text)

    # stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]


    preprocessed_text = ' '.join(filtered_tokens)
    return preprocessed_text

def remove_numbers(text):
    text=''.join([i for i in text if not i.isdigit()])
    return text
```

```python
df_cyber['post_tokens'] = df_cyber['post_tokens'].apply(preprocess_text)
```

```python
df_cyber['post_tokens'] = df_cyber['post_tokens'].apply(preprocess_text)
```

**Dataset and DataLoader**

```python
class CyberBullyingDataset(Dataset):
    def __init__(self, input_ids, attention_mask, labels):
        self.input_ids = input_ids
        self.attention_mask = attention_mask
        self.labels = labels #indicates whether the post is cyberbullying or not

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            'input_ids': self.input_ids[idx],
            'attention_mask': self.attention_mask[idx],
            'labels': self.labels[idx]
        }
```

The code defines a custom Pytorch dataset called Cyberbullying which will handle and organize input data for a cyberbullying detection model.

It stores the tokenized inputs, attention_mask and corresponding labels.

The class allows for efficient indexing and retrieval of data samples during model training.  

```python
label_tensor = torch.tensor(df_cyber['label_encoder'].values)
dataset = CyberBullyingDataset(input_ids,attention_mask, label_tensor)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

The dataset - This object now holds the processed data which includes the tokenized post, attention_masks and labels

Dataloader - training the model efficiently

**Model Making for BERT classification**

```python
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu') # to speed up the training process of machine learning models
model.to(device)
```

```python
optimizer = AdamW(model.parameters(), lr=2e-5, eps=1e-8)
num_epochs = 15
# Total number of training steps
total_steps = len(dataloader) * num_epochs

scheduler = get_linear_schedule_with_warmup(optimizer,
                                            num_warmup_steps=0,
                                            num_training_steps=total_steps)
```

Adamw - A optimizer that adjusts the model parameters based on the error it makes and also it helps the model to prevent from overfitting.

lr = 2e-5 - This sets the learning rate which controls how much the optimizer should change the model's parameters.

eps=1e-8: This is a tiny number added to prevent division by zero errors during optimization. It ensures numerical stability.

get_linear_schedule_with_warmup: This function creates a schedule where the learning rate starts at zero, gradually increases to the initial learning rate, and then linearly decreases to zero by the end of training

```python
epochs = 10  # looks at the entire dataset 10 times

for epoch in range(epochs):
    model.train()
    total_loss = 0 #(how far the model's prediction are from actual labels)
    correct_predictions = 0 # it counts how many prediction the models get correct during epoch

    for batch in dataloader:
        optimizer.zero_grad() # cleaning previous gradients to ensure that model updates correctly based on current batch

        #Moving the data to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)


        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss # The loss is calculated based on how far off the model's predictions are from the true labels
        logits = outputs.logits# These are the raw predictions made by the model before applying any activation function


        total_loss += loss.item()

        #BACKPROPOGATION AND OPTIMIZATION
        loss.backward()
        optimizer.step()
        scheduler.step()


        preds = torch.argmax(F.softmax(logits, dim=1), dim=1)# Converts the model's raw predictions (logits) into final class predictions
        correct_predictions += torch.sum(preds == labels)# Counts how many of the model's predictions match the true labels
                                                        # and adds them to the total number of correct predictions.

    avg_loss = total_loss / len(dataloader)
    accuracy = correct_predictions.double() / len(dataloader.dataset)

    print(f'Epoch {epoch+1}, Loss: {avg_loss}, Accuracy: {accuracy}')
```

Loss represents how wrong the model's prediction are. If lower loss means the model is doing great and the goal is to minimize loss during training.

```python
test_comments = df_final_cyber['comment'].tolist()

test_encodings = tokenizer(test_comments, padding=True, truncation=True, return_tensors='pt')

test_labels = torch.tensor(df_final_cyber['label_encoder'].values)
```

```python
test_dataset = CyberBullyingDataset(test_encodings['input_ids'], test_encodings['attention_mask'], test_labels)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
model.eval()
predictions = []
true_labels = []

with torch.no_grad():
    for batch in test_dataloader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        # Forward pass
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits


        preds = torch.argmax(logits, dim=1)
        predictions.extend(preds.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(true_labels, predictions)
print(f'The accuracy score : {accuracy:.4f}')
```

The model achieved 93.04% accuracy, meaning it correctly classified cyberbullying posts 93 out of 100 times

**Summary**

In the cyberbullying detection project, BERT is used to understand the meaning of social media comments and classify whether they contain cyberbullying based on categories like race, gender, or religion.

How BERT works in this project:

Text Input: BERT takes the social media comments (text) and tokenizes them (splits them into words and subwords that the model can understand).

Contextual Understanding: BERT looks at each word in the comment in relation to other words to understand the full context. For example, it can tell the difference between "run fast" and "fast food," where "fast" has different meanings.

Classification: After processing the comment, BERT outputs a prediction: whether the comment is bullying or not, and if so, which category (e.g., based on race, gender, etc.).

Real-world example:

Imagine someone posts: "You're such a loser, nobody likes you because of where you're from."

BERT looks at the entire sentence and understands that this message might involve cyberbullying related to race or nationality.
After processing the text, BERT predicts this as cyberbullying in the "race" category.
In simple terms, BERT helps detect harmful or offensive comments by analyzing the words and their context in the sentence to determine if it's bullying and the type of bullying involved.
